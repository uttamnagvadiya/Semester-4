```python
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]


# write a code to generate below graph
plt.plot(x, cxMarks, color='r', ls='--', marker='o', )
plt.plot(x, cyMarks, color='b', ls=':', marker='>', )
plt.xlabel("Roll No")
plt.ylabel("CPI")
plt.annotate(xy=[8,1], text="Lowest SPI")
plt.legend(["cx", "cy"], loc=3)
plt.show()
```

### 01) WAP to demonstrate the use of Pie chart.

```python
x = [10,20,30,40]
y = ["Food", "Travel", "Acomodation", "Shopping"]
c = ['r', 'b', 'y', 'g']
ex = [0.5,0.0,0.0,0.0]

plt.pie(x, labels=y, colors=c, explode=ex, shadow=True, radius=2, autopct="%0.2f%%", labeldistance=1.3, textprops={"fontsize":20})

plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x = ["python", "c", "c++", "java"]
y = [85,70,60,80]
c = ['r', 'b', 'y', 'g']

plt.barh(x, y, color=c, edgecolor='b', alpha=0.7)
plt.title("Popularity of Languages")
plt.ylabel("Languages")
plt.xlabel("Rating")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = ["python", "c", "c++", "java"]
y = [85,70,60,80]
c = ['r', 'b', 'y', 'g']

plt.bar(x, y, color=c, edgecolor='b', alpha=0.7)
plt.title("Popularity of Languages")
plt.xlabel("Languages")
plt.ylabel("Rating")

def addLabels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i])
        
addLabels(x, y)

# plt.show()
plt.savefig("mychart.png")
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x = [1,3,5,7,9]
y = [2,4,6,8,10]
c = ['r', 'b', 'y', 'g', 'k']
size = [100,200,300,400,500]
plt.scatter(x, y, color=c, marker='*', s=size)
plt.title("Scatter Plot")
plt.xlabel("Odd")
plt.ylabel("Even")
plt.show()
```
